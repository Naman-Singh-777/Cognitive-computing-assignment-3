```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("Q1: Matrix Operations")
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print("Sum of all elements:", np.sum(gfg))
print("Sum of elements row-wise:", np.sum(gfg, axis=1))
print("Sum of elements column-wise:", np.sum(gfg, axis=0))
```

```python
print("\nQ2(a): Array Operations")
array = np.array([10, 52, 62, 16, 16, 54, 453])
sorted_array = np.sort(array)
print("Sorted array:", sorted_array)
print("Indices of sorted array:", np.argsort(array))
print("4 smallest elements:", sorted_array[:4])
print("5 largest elements:", sorted_array[-5:])
```

```python
print("\nQ2(b): Float and Integer Operations")
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print("Integer elements:", array[array == array.astype(int)])
print("Float elements:", array[array != array.astype(int)])
```

```python
print("\nQ3: Sales Data Operations")
X = ord('N') + ord('S')
sales = np.array([X, X+50, X+100, X+150, X+200])
print("Original sales:", sales)

tax_rate = ((X % 5) + 5) / 100
taxed_sales = sales * (1 + tax_rate)
print("Sales after tax:", taxed_sales)

discount_mask = sales < X+100
discounted_sales = np.where(discount_mask, sales * 0.95, sales * 0.90)
print("Sales after discount:", discounted_sales)

weekly_sales = np.tile(sales, (3, 1))
weekly_increase = np.array([1, 1.02, 1.04]).reshape(-1, 1)
final_sales = weekly_sales * weekly_increase
print("Weekly sales with 2% increase:", final_sales)
```

```python
print("\nQ4: Function Plotting")
x = np.linspace(-10, 10, 100)
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x, x**2)
plt.title('Y = x^2')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 2)
plt.plot(x, np.sin(x))
plt.title('Y = sin(x)')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 3)
plt.plot(x, np.exp(x))
plt.title('Y = e^x')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(2, 2, 4)
plt.plot(x, np.log(np.abs(x) + 1))
plt.title('Y = log(|x| + 1)')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
plt.show()
```
